```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 20)
```

```python
y = np.random.randint(0, 50, 20)
```

```python
plt.plot(x, y, color='g', linestyle='--', marker='o', markersize=5, markerfacecolor='red')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y axis')
plt.show()
```

```python
y_views = np.random.rand(7)
f_views = np.random.rand(7)
t_views = np.random.rand(7)
days = np.arange(1, 8)
```

```python
plt.scatter(days, y_views, label='Youtube Views', marker='x')
plt.scatter(days, f_views, label='Facebook Views', marker='o')
plt.scatter(days, t_views, label='Twitter Views', marker='o')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media Views')
plt.grid(linestyle='--', color='r')
plt.legend()
```

```python
plt.bar(days, y_views, label='Youtube Views')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media Views')
plt.legend()
```
