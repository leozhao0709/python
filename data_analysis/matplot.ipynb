```python
t = np.arange(0, 2, 0.1)
```

```python
s = np.sin(t*np.pi)
```

```python
plt.plot(t, s, 'r--', label='aaaa') # red line with label aaaa
plt.plot(t*2, s, 'b--', label='bbbb') # blue line with label bbbb
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('this is title')
plt.legend()
```

```python
plt.subplot(2, 1, 1)
plt.plot(t, s, 'r--')
plt.ylabel('y1')
plt.subplot(2, 1, 2)
plt.plot(t, 2*s, 'b--')
plt.ylabel('y2')
plt.xlabel('x')
```
